```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def Graph_without_Seperator( x ):

    if( isinstance( x, np.ndarray ) == False ):
                    x = np.array( x )
    
    row, col = x.shape
    
    columns = []
    columns_bool = []
    
    for i in range( col ):
        columns.append( x[ :, i ] )
        columns_bool.append( x[ :, i ] > 0 )
    
    columns = np.array( columns )
    columns_bool = np.array( columns_bool )

    y_xor = []

    for i in range( len( columns_bool[0] ) ):

        if( ( columns_bool[0][i] == True and columns_bool[1][i] == True ) or 
            ( columns_bool[0][i] == False and columns_bool[1][i] == False ) ):
            
            y_xor.append( False )

        else:
            y_xor.append( True )

    y_xor = np.array( y_xor )
    
    y_xor = np.where(y_xor, 1, -1)
    
    minus_x = []
    plus_x = []
    
    print( x )
    
    for i in range( col ):
        plus_x.append( x[y_xor == 1, i] )
        minus_x.append( x[y_xor == -1, i] )
    
    plt.scatter(plus_x[0], plus_x[1],
            color = 'blue', marker='x',
            label='1')

    plt.scatter(minus_x[0], minus_x[1],
                color='red',
                marker='o',
                label='-1')

    plt.xlim([-3, 3])
    plt.ylim([-3, 3])

    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
Graph_without_Seperator(np.random.randn(200, 2) )
```
